```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import requests
```

```python
url="https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
```

```python
r = requests.get(url, headers= 0)
tables = pd.read_html(r.text)
```

```python
df=pd.DataFrame(tables[0])
df.columns=['Postcode','Borough','Neighbourhood']

df.drop([0],axis=0,inplace=True)

df.reset_index()
```

```python
filtered_df = df[df.Borough != "Not assigned"]
filtered_df.shape
```

```python
filtered_df['New_Neighbourhood'] = np.where(filtered_df['Neighbourhood']=='Not Assigned', filtered_df['Borough'], filtered_df['Neighbourhood'])
filtered_df.head(10)
```

```python
postal_df = pd.DataFrame(filtered_df.groupby(['Postcode','Borough'])['New_Neighbourhood'].apply('.'.join).reset_index())
```

```python
postal_df = postal_df.rename(columns = {"Postcode": "PostalCode","New_Neighbourhood":"Neighbourhood"})
print(postal_df.head())
```

```python
postal_df.shape
```

```python
CanadaLatLon = pd.read_csv('https://cocl.us/Geospatial_data')
CanadaLatLon.head()
```

```python


CanadaLatLon.rename(columns ={'Postal Code':'PostalCode'}, inplace = True)
Canada_df = pd.merge(postal_df, CanadaLatLon, on='PostalCode')
Canada_df.head()

```

```python
Toronto_df = Canada_df[Canada_df['Borough'].str.contains('Toronto',regex = False)]
Toronto_df.head()
```
